```python
#Z-Score Normalization
import pandas as pd
import numpy as np
import math
a=np.array([1,21,34,57,76,45,97,100,102])
b=[]
sum=0

#mean calc
for i in range(len(a)):
    sum=sum+a[i]
mean=sum/len(a)

#varience calc
var=0
for i in range(len(a)):
    var=var+(a[i]-mean)**2
varience=var/len(a)

#standard deviation calc
std=math.sqrt(varience)

#normalize
for i in range(len(a)):
    b.append((a[i]-mean)/std)
print(b)
```

```python
import numpy as np

a = [9,8,7,6,np.nan,5,4,3,2,1,0]

# Step 1: sort the list ignoring NaN
b = [x for x in a if x == x]   # keep only non-NaN values
b.sort()

# Step 2: find median
n = len(b)
if n % 2 != 0:
    median = b[n // 2]
else:
    median = (b[n // 2 - 1] + b[n // 2]) / 2

print("Median:", median)

# Step 3: replace NaN with median
for i in range(len(a)):
    if a[i] != a[i]:   # NaN check
        print("None value stored in the index", i)
        a[i] = median

print(a)
```
