```python
M = 5  
N = 2   
K = 3  

D = [
    [1, 2],
    [2, 3],
    [3, 1],
    [6, 5],
    [7, 7]
]

C = [0, 0, 0, 1, 1]

Q = [4, 3]

dist = [0] * M

for i in range(M):              
    sum_val = 0
    for j in range(N):          
        sum_val += abs(D[i][j] - Q[j])
    dist[i] = sum_val / N       

for i in range(M - 1):
    for j in range(M - i - 1):
        if dist[j] > dist[j + 1]:
            temp = dist[j]
            dist[j] = dist[j + 1]
            dist[j + 1] = temp

            temp_class = C[j]
            C[j] = C[j + 1]
            C[j + 1] = temp_class

count0 = 0
count1 = 0

for i in range(K):
    if C[i] == 0:
        count0 += 1
    else:
        count1 += 1

if count0 > count1:
    print("Predicted Class = 0")
else:
    print("Predicted Class = 1")
```
